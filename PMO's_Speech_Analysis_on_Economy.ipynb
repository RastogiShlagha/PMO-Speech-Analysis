Using Beautiful Soup

```
# Installing all the required libraries -
# !pip install requests
# !pip install beautifulsoup4
# !pip install googletrans
url = "https://www.pmindia.gov.in/en/tag/pmspeech/"
```

```
# Function to parse through the main website
import requests
from bs4 import BeautifulSoup

def html_parser(url):
  page = requests.get(url) # Requesting to fetch URL
  # print(page)
  
  html_page = BeautifulSoup(page.content, 'html.parser') # Converting HTML to text
  # print(html_page)

  return html_page
```

```
basefilter = html_parser(url).find(id='wrapper').find_all('li', class_='6u')
# print(basefilter.prettify())

doc = []
link_list = []

from googletrans import Translator
translator = Translator()

for tag in basefilter:
    speechlink_tag = tag.find_all('h3') # Fetching all h3 tags
    # print(speechlink_tag)

    for tags in speechlink_tag:
      link = tags.find('a')['href'] # Fetching all href links
      
      link_parser = html_parser(link) # Parsing each href link to text
      speechlink_filter = link_parser.find_all('div', class_="news-bg") # Fetching all main data from div tags
      
      for x in speechlink_filter :
        xx = x.text.rstrip("\n") # Fetching text from main tag, with newline removed
        xxx = translator.translate(xx, src = 'hi', dest = 'en') # Translating hindi text(if any) to english

        # print(xxx.text)
        doc.append(xxx.text)

      link_list.append(link)
      # print(f"Speech link here : {link}\n")
```

```
for i in range(len(link_list)):
  print(link_list[i], doc[i])
  print()
```

```
# Finding the frequency of our word -  Economy

def freq(str): 
  # break the string into list of words 
  str_list = str.split() 

  # gives set of unique words 
  unique_words = set(str_list) 
    
  for words in unique_words : 
      print('Frequency of ', words , 'is :', str_list.count(words))

from lxml import html

page = requests.get(url)
tree = html.fromstring(page.content)
dates = tree.xpath('//span[@class="date"]/text()')

for i in range(len(doc)):
  print('Count of Economy in PMOs speech, dated', dates[i], 'is :', doc[i].count('economy'))
  # print (doc[i])
  print()
```
